**Introduction to Pandas**

```
# import pandas library
# pd is the short name used for pandas

import pandas as pd

print("Pandas imported successfully")
```

**What is pandas?**

```
# pandas is used to work with table data
# like Excel table or CSV file

print("Pandas helps us work with rows and columns")
```

**Creating a Series**

```
# Series = one single column of data

s = pd.Series([10, 20, 30, 40, 50])

print(s)

# Series is like one column
# index is on left
# values are on right
```

```
# Series with custom index

# we can also give our own labels/index

s2 = pd.Series([90, 85, 88], index=["Math", "English", "Science"])

print(s2)

# custom labels make data more meaningful
```

Create a DataFrame

```
# DataFrame = table with rows and columns

data = {
    "Name": ["Madhu", "Rahul", "Priya", "Arjun", "Neha"],
    "Age": [20, 21, 19, 22, 20],
    "Course": ["MSC", "BSc", "BBA", "BCA", "BSc"],
    "Marks": [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

print(df)

# dictionary keys become column names
# each list becomes one column
# DataFrame is the main pandas object
```

```
# Check type

# type() tells what kind of object it is

print(type(df))

# this confirms it is a DataFrame
```

```
# View first rows

# head() shows first 5 rows by default

df.head()

# useful for quickly checking big datasets
```

```
# View last rows

# tail() shows last 5 rows by default

df.tail()
```

```
# Column names

# columns shows all column names

df.columns
```

```
# Basic info

# info() gives summary
# shows column names, data types, non-null values

df.info()

# very useful for understanding dataset structure
```

```
# Statistical summary

# describe() gives summary of numerical columns

df.describe()
```

```
# Select one column

# select one column using column name

# df["Name"]

# Select multiple columns

# select multiple columns using list

df[["Name", "Marks"]]
```

```
# Select a row using index

# loc is used to select row by label/index

df.loc[0]

# row 0 means first row
```

```
# Select multiple rows

# select rows 0 to 2

df.loc[0:2]

# in loc, end value is included
```

```
# Select row and specific columns

# select row 1 and 2 with Name and Marks columns

df.loc[1:2, ["Name", "Marks"]]
```

```
# # iloc selects by position number

df.iloc[0]

# loc = label based
# iloc = position based
```

```
# Filter data

# show students with marks greater than 80

df[df["Marks"] > 85]

# filtering means showing only matching rows
```

```
# Another filter

# show students whose course is BCA

df[df["Course"] == "MSC"]
```

```
# Multiple conditions

# use & for AND condition
# use brackets around each condition

df[(df["Marks"] > 80) & (df["Course"] == "BSc")]

# & means both conditions must be true
```

```
# Add a new column

# create a new column called Pass
# if marks >= 80 then pass = Yes else No

df["Pass"] = ["Yes", "No", "Yes", "No", "Yes"]

df
```

```
# Add another calculated column

# add 5 grace marks to every student

df["Updated_Marks"] = df["Marks"] + 5

df
```
